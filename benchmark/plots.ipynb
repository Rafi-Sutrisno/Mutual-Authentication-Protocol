```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSVs
rsa = pd.read_csv("results\\output_rsa_stats_history.csv")
rsa_puf = pd.read_csv("results\\output_puf_stats_history.csv")

# Normalize timestamp
rsa['Time'] = rsa['Timestamp'] - rsa['Timestamp'].iloc[0]
rsa_puf['Time'] = rsa_puf['Timestamp'] - rsa_puf['Timestamp'].iloc[0]

# ---------------------------
# 1. Requests per second
# ---------------------------
plt.figure(figsize=(10,5))
plt.plot(rsa['Time'], rsa['Requests/s'], label="RSA")
plt.plot(rsa_puf['Time'], rsa_puf['Requests/s'], label="RSA+PUF")
plt.xlabel("Time (s) since start")
plt.ylabel("Requests per second")
plt.title("Throughput Over Time")
plt.legend()
plt.grid(True)
plt.show()

# ---------------------------
# 2. Average Response Time
# ---------------------------
plt.figure(figsize=(10,5))
plt.plot(rsa['Time'], rsa['Total Average Response Time'], label="RSA")
plt.plot(rsa_puf['Time'], rsa_puf['Total Average Response Time'], label="RSA+PUF")
plt.xlabel("Time (s) since start")
plt.ylabel("Average Response Time (ms)")
plt.title("Average Response Time Over Time")
plt.legend()
plt.grid(True)
plt.show()

# ---------------------------
# 3. 95th Percentile Response Time
# ---------------------------
plt.figure(figsize=(10,5))
plt.plot(rsa['Time'], rsa['95%'], label="RSA 95th %")
plt.plot(rsa_puf['Time'], rsa_puf['95%'], label="RSA+PUF 95th %")
plt.xlabel("Time (s) since start")
plt.ylabel("95th Percentile Response Time (ms)")
plt.title("95th Percentile Response Time Over Time")
plt.legend()
plt.grid(True)
plt.show()

# ---------------------------
# 4. Failures per second
# ---------------------------
plt.figure(figsize=(10,5))
plt.plot(rsa['Time'], rsa['Failures/s'], label="RSA")
plt.plot(rsa_puf['Time'], rsa_puf['Failures/s'], label="RSA+PUF")
plt.xlabel("Time (s) since start")
plt.ylabel("Failures per second")
plt.title("Failures Over Time")
plt.legend()
plt.grid(True)
plt.show()

# ---------------------------
# 5. Concurrent Users
# ---------------------------
plt.figure(figsize=(10,5))
plt.plot(rsa['Time'], rsa['User Count'], label="RSA")
plt.plot(rsa_puf['Time'], rsa_puf['User Count'], label="RSA+PUF")
plt.xlabel("Time (s) since start")
plt.ylabel("Number of Users")
plt.title("Concurrent Users Over Time")
plt.legend()
plt.grid(True)
plt.show()
```
